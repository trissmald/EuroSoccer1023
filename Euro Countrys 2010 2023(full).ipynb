# England Leagues 2010 - 2023

```python
# Analisis de la estructura del link.
# https://www.football-data.co.uk/mmz4281/2223/E0.csv: "https://www.football-data.co.uk/mmz4281/" + "2223" + "/" + "E0" + ".csv"

# Nomenglatura de ligas:
# Premier League (FT & HT results; match stats; match, total goals & AH odds) / E0
# Championship (FT & HT results; match stats; match, total goals & AH odds) / E1
# League 1 (FT & HT results; match stats; match, total goals & AH odds) / E2
# League 2 (FT & HT results; match stats; match, total goals & AH odds) / E3
```

```python
# Declaramos una variable root con la parte del enlace que no cambia.
root_e = "https://www.football-data.co.uk/mmz4281/"
```

```python
# Listas para almancenar las ligas
leagues = ['E0', 'E1', 'E2', 'E3']
england = [] # Aquí se almacenarán los nuevos data frames

# Ciclo para leer los multiples cvs y adjuntarlos.
for league in leagues:
    for season in range(10, 23):
        df = pd.read_csv(root_e + str(season) + str(season+1) + '/' + league + '.csv', encoding='unicode_escape') #encoding para no generar errores en la compilación
        df.insert(1, 'Season', season)
        df.rename(columns={'Div': 'League'}, inplace=True)
        df['League'].replace('E0', 'Premier League', inplace=True)
        df['League'].replace('E1', 'Championship', inplace=True)
        df['League'].replace('E2', 'League 1', inplace=True)
        df['League'].replace('E3', 'League 2', inplace=True)
        england.append(df) # Forma de almacenar los DF que se van a iterar en frames.
```

```python
len(england) #contamos los elementos que son 12 temporadas X 4 Ligas desde
```

```python
england[0]
```

# Spain Leagues 2010 - 2023

```python
# Analisis de la estructura del link.
# https://www.football-data.co.uk/mmz4281/2223/E0.csv: "https://www.football-data.co.uk/mmz4281/" + "2223" + "/" + "SP1" + ".csv"

# Nomenglatura de ligas:
# La Liga Primera Division (FT & HT results; match stats; match, total goals & AH odds) / SP1
# La Liga Segunda Division (FT & HT results; match stats; match odds and total goals odds) / SP2
```

```python
# Declaramos una variable root con la parte del enlace que no cambia.
root_sp= "https://www.football-data.co.uk/mmz4281/"
```

```python
# Listas para almancenar las ligas
leagues = ['SP1', 'SP2']
spain = [] # Aquí se almacenarán los nuevos data frames
league_to_country = {'SP1': 'Spain', 
                     'SP2': 'Spain'}

# Ciclo para leer los multiples cvs y adjuntarlos.
for league in leagues:
    for season in range(10, 23):
        df = pd.read_csv(root_sp + str(season) + str(season+1) + '/' + league + '.csv', encoding='unicode_escape') #encoding para no generar errores en la compilación
        df.insert(1, 'Season', season)
        df.rename(columns={'Div': 'League'}, inplace=True)
        df['Country'] = df['League'].map(league_to_country)
        df.insert(0, 'Country', df.pop('Country'))
        

        columnas_mantener = ['Date', 'Season', 'League', 'Country', 'Time', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR', 'HTHG', 'HTAG', 'HTR', 'Attendance', 'Referee', 'HS', 'AS', 'HST', 'AST', 'HHW', 'AHW', 'HC', 'AC', 'HF', 'AF', 'HFKC', 'AFKC', 'HO', 'AO', 'HY', 'AY', 'HR', 'AR', 'HBP', 'ABP']
        df.drop(columns=[col for col in df.columns if col not in columnas_mantener], inplace=True)
       
        df['League'].replace('SP1', 'La Liga 1', inplace=True)
        df['League'].replace('SP2', 'La Liga 2', inplace=True)

        spain.append(df) # Forma de almacenar los DF que se van a iterar en frames.
        
```

```python
len(spain)
```

```python
spain[10]
```
